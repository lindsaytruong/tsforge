## Plot Decomposition

```python
from __future__ import annotations

import warnings
from dataclasses import dataclass
from typing import Iterable, Literal, Optional

import numpy as np
import pandas as pd

Method = Literal["stl", "rstl", "mstl", "fourier", "classical"]
Model = Literal["additive", "multiplicative"]


@dataclass
class DecomposeConfig:
    method: Method = "stl"
    period: Optional[int] = None
    seasonal_periods: Optional[Iterable[int]] = None
    model: Model = "additive"
    robust: bool = True
    lowess_frac: float = 0.15
    fourier_order: int = 3
    freq: Optional[str] = None
    enforce_sorted: bool = True
    return_seasonal_sum: bool = True
    intermittent_threshold: Optional[float] = 1.32
    min_obs: Optional[int] = None  # auto: 2*max(periods)
    verbose: bool = False          # Nixtla-style: silent by default


# --------------------------
# Helper functions
# --------------------------

def _validate_regular_or_warn(g: pd.DataFrame, date_col: str, freq: Optional[str]) -> None:
    if freq is None:
        return
    idx = pd.DatetimeIndex(g[date_col].values)
    inferred = pd.infer_freq(idx)
    if inferred is None or (inferred != freq):
        warnings.warn(
            f"[decompose_series] Irregular cadence; expected {freq}, inferred {inferred}.",
            UserWarning,
        )


def _stl_decompose(y: pd.Series, period: int, robust: bool):
    from statsmodels.tsa.seasonal import STL
    mask = y.notna()
    out = pd.DataFrame(index=y.index, columns=["trend", "seasonal", "remainder"], dtype=float)
    if mask.sum() < max(3, period + 2):
        return out
    stl = STL(y[mask], period=period, robust=robust)
    res = stl.fit()
    out.loc[mask, "trend"] = res.trend
    out.loc[mask, "seasonal"] = res.seasonal
    out.loc[mask, "remainder"] = res.resid
    return out


def _classical_decompose(y: pd.Series, period: int, model: Model):
    from statsmodels.tsa.seasonal import seasonal_decompose
    mask = y.notna()
    out = pd.DataFrame(index=y.index, columns=["trend", "seasonal", "remainder"], dtype=float)
    if mask.sum() < max(3, period + 2):
        return out
    res = seasonal_decompose(y[mask], model=model, period=period, two_sided=True, extrapolate_trend="freq")
    out.loc[mask, "trend"] = res.trend
    out.loc[mask, "seasonal"] = res.seasonal
    out.loc[mask, "remainder"] = res.resid
    return out


def _mstl_decompose(y: pd.Series, seasonal_periods: Iterable[int], robust: bool, return_sum: bool = True):
    try:
        from statsmodels.tsa.seasonal import MSTL
    except Exception as e:
        raise ImportError("statsmodels>=0.14 is required for MSTL.") from e

    mask = y.notna()
    seas_periods = list(seasonal_periods)
    seas_cols = [f"seasonal_{p}" for p in seas_periods]

    out = pd.DataFrame(index=y.index, columns=["trend", "remainder"] + seas_cols, dtype=float)
    if mask.sum() < (max(seas_periods) + 3):
        return out

    import inspect
    sig = inspect.signature(MSTL.__init__)
    kwargs = {"periods": seas_periods}
    if "robust" in sig.parameters:
        kwargs["robust"] = robust

    mstl = MSTL(y[mask], **kwargs)
    res = mstl.fit()

    out.loc[mask, "trend"] = res.trend
    out.loc[mask, "remainder"] = res.resid

    comp = getattr(res, "seasonal_components", None)
    if comp is not None:
        comp = np.asarray(comp)
        for j, p in enumerate(seas_periods):
            out.loc[mask, f"seasonal_{p}"] = comp[:, j]
    else:
        seas = getattr(res, "seasonal", None)
        if seas is not None:
            if isinstance(seas, pd.DataFrame):
                for j, p in enumerate(seas_periods):
                    if j < seas.shape[1]:
                        out.loc[mask, f"seasonal_{p}"] = seas.iloc[:, j].values
            else:
                out.loc[mask, f"seasonal_{seas_periods[0]}"] = seas.values

    if return_sum:
        seas_available = [c for c in seas_cols if out[c].notna().any()]
        if seas_available:
            out["seasonal"] = out[seas_available].sum(axis=1)

    return out


def _fourier_design(n: int, periods: Iterable[int], order: int) -> np.ndarray:
    t = np.arange(n, dtype=float)
    cols = []
    for P in periods:
        for k in range(1, order + 1):
            cols.append(np.sin(2.0 * np.pi * k * t / P))
            cols.append(np.cos(2.0 * np.pi * k * t / P))
    return np.column_stack(cols) if cols else np.zeros((n, 0), dtype=float)


def _fourier_decompose(y: pd.Series, seasonal_periods: Iterable[int], fourier_order: int, lowess_frac: float):
    from statsmodels.nonparametric.smoothers_lowess import lowess
    idx = y.index
    mask = y.notna()
    n = mask.sum()
    out = pd.DataFrame(index=idx, columns=["trend", "seasonal", "remainder"], dtype=float)
    if n < 5:
        return out
    yy = y[mask].astype(float).values
    x = np.arange(n, dtype=float)
    trend_vals = lowess(yy, x, frac=lowess_frac, it=1, return_sorted=False)
    detrended = yy - trend_vals
    periods = list(seasonal_periods) if seasonal_periods else []
    X = _fourier_design(n, periods, fourier_order)
    if X.shape[1] > 0:
        Xreg = np.column_stack([np.ones(n), X])
        beta, _, _, _ = np.linalg.lstsq(Xreg, detrended, rcond=None)
        seasonal_vals = Xreg @ beta
        intercept = beta[0]
        trend_vals = trend_vals + intercept
        seasonal_vals = seasonal_vals - intercept
    else:
        seasonal_vals = np.zeros_like(yy)
    remainder_vals = yy - trend_vals - seasonal_vals
    out.loc[mask, "trend"] = trend_vals
    out.loc[mask, "seasonal"] = seasonal_vals
    out.loc[mask, "remainder"] = remainder_vals
    return out


# --------------------------
# Main function
# --------------------------

def decompose_series(
    df: pd.DataFrame,
    id_col: str,
    date_col: str,
    value_col: str,
    *,
    method: Method = "stl",
    period: Optional[int] = None,
    seasonal_periods: Optional[Iterable[int]] = None,
    model: Model = "additive",
    robust: bool = True,
    lowess_frac: float = 0.15,
    fourier_order: int = 3,
    freq: Optional[str] = None,
    enforce_sorted: bool = True,
    return_seasonal_sum: bool = True,
    intermittent_threshold: Optional[float] = 1.32,
    min_obs: Optional[int] = None,
    verbose: bool = False,
) -> pd.DataFrame:
    """
    Decompose one or many time series into trend / seasonal / remainder.

    Adds:
      - ADI
      - Flags: series_flag ('regular','intermittent','short','intermittent+short')
      - Booleans: is_intermittent, is_short
      - Skips decomposition if flagged
    """
    out_frames = []
    df_work = df[[id_col, date_col, value_col]].copy()
    df_work[date_col] = pd.to_datetime(df_work[date_col])
    if enforce_sorted:
        df_work = df_work.sort_values([id_col, date_col])

    for sid, g in df_work.groupby(id_col, observed=True):
        if enforce_sorted:
            g = g.sort_values(date_col)

        y = g[value_col].astype(float)
        n_obs = len(y)
        n_nonzero = (y > 0).sum()
        adi = n_obs / n_nonzero if n_nonzero > 0 else np.inf

        # --- compute required min length ---
        if min_obs is not None:
            required = min_obs
        else:
            longest = 0
            if period:
                longest = max(longest, period)
            if seasonal_periods:
                longest = max(longest, max(seasonal_periods))
            required = max(10, 2 * longest) if longest > 0 else 10

        # --- check conditions ---
        is_intermittent = intermittent_threshold is not None and adi > intermittent_threshold
        is_short = n_obs < required

        # series_flag with combos
        if is_intermittent and is_short:
            flag = "intermittent+short"
        elif is_intermittent:
            flag = "intermittent"
        elif is_short:
            flag = "short"
        else:
            flag = "regular"

        # --- skip decomposition if flagged ---
        if flag != "regular":
            if verbose:
                warnings.warn(
                    f"Series {sid} flagged '{flag}' "
                    f"(ADI={adi:.2f}, n_obs={n_obs}, required={required}). "
                    "Skipping decomposition.",
                    UserWarning,
                )
            comp = pd.DataFrame(index=y.index, columns=["trend","seasonal","remainder"], dtype=float)
            comp["trend"] = 0.0
            comp["seasonal"] = 0.0
            comp["remainder"] = y
        else:
            _validate_regular_or_warn(g, date_col, freq)
            method_l = method.lower()
            if method_l == "stl":
                comp = _stl_decompose(y, period=period, robust=robust)
            elif method_l == "rstl":
                comp = _stl_decompose(y, period=period, robust=True)
            elif method_l == "classical":
                comp = _classical_decompose(y, period=period, model=model)
            elif method_l == "mstl":
                comp = _mstl_decompose(y, seasonal_periods=seasonal_periods, robust=robust,
                                       return_sum=return_seasonal_sum)
            elif method_l == "fourier":
                comp = _fourier_decompose(y, seasonal_periods=seasonal_periods,
                                          fourier_order=fourier_order, lowess_frac=lowess_frac)
            else:
                raise ValueError(f"Unknown method: {method}")

        # --- assemble output ---
        base = pd.DataFrame({id_col: sid, date_col: g[date_col], "observed": y.values})
        for col in comp.columns:
            base[col] = comp[col].values
        base["ADI"] = adi
        base["series_flag"] = flag
        base["is_intermittent"] = is_intermittent
        base["is_short"] = is_short
        out_frames.append(base)

    result = pd.concat(out_frames, axis=0, ignore_index=True)

    # tidy columns
    cols = [id_col, date_col, "observed", "ADI", "series_flag", "is_intermittent", "is_short"]
    for c in ["trend", "seasonal", "remainder"]:
        if c in result.columns:
            cols.append(c)
    seas_cols = [c for c in result.columns if c.startswith("seasonal_")]
    cols.extend(sorted(seas_cols))
    for c in result.columns:
        if c not in cols:
            cols.append(c)
    return result[cols]
```

```python
import pandas as pd
import numpy as np
from typing import Union, List, Optional, Literal
from tsforge.plots.style import _apply_tsforge_style, PALETTE, HIGHLIGHT


def plot_decomposition(
    df: pd.DataFrame,
    id_col: str,
    date_col: str,
    ids: Union[None, int, str, List[str]] = None,
    engine: Literal["plotly", "matplotlib"] = "plotly",
    mode: Literal["overlay", "facet", "dropdown"] = "overlay",
    title: Optional[str] = None,
):
    """
    Plot decomposition components returned by `decompose_series`.

    Assumes `df` contains columns like:
      observed, trend, seasonal (optional), seasonal_<period> (optional), remainder

    Parameters
    ----------
    df : DataFrame
        Output from decompose_series(...) including [id_col, date_col] and components.
    id_col : str
        Series identifier column.
    date_col : str
        Timestamp column (datetime).
    ids : None | int | str | list, default None
        - None: sample up to 6 series.
        - int: sample that many series.
        - str: single series.
        - list[str]: explicit series list.
    engine : {'plotly','matplotlib'}, default 'plotly'
        Plotting backend.
    mode : {'overlay','facet','dropdown'}, default 'overlay'
        - overlay  : draw selected series together (lines stacked by component row).
        - facet    : same as overlay but intended for a few series; (set ids=1 for single).
        - dropdown : dropdown switches which series is shown (plotly only).
    title : str, optional
        Custom title; default is 'Decomposition'.

    Returns
    -------
    Figure
        plotly.graph_objs.Figure or matplotlib.figure.Figure
    """

    # ---------- prep ----------
    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col])

    # Pick IDs
    unique_ids = df[id_col].dropna().unique().tolist()
    if ids is None:
        n_series = min(6, len(unique_ids))
        ids = pd.Series(unique_ids).sample(n_series, random_state=42).tolist()
    elif isinstance(ids, int):
        n_series = min(ids, len(unique_ids))
        ids = pd.Series(unique_ids).sample(n_series, random_state=42).tolist()
    elif isinstance(ids, str):
        ids = [ids]
        n_series = 1
    else:
        ids = list(ids)
        n_series = len(ids)

    df = df[df[id_col].isin(ids)].copy()
    df = df.sort_values([id_col, date_col])

    # Detect components
    cols = df.columns.tolist()
    has_observed = "observed" in cols
    has_trend = "trend" in cols
    has_seasonal_sum = "seasonal" in cols
    has_remainder = "remainder" in cols
    seasonal_parts = [c for c in cols if c.startswith("seasonal_")]  # e.g., seasonal_7, seasonal_365

    # Assemble component order (rows)
    component_rows: List[str] = []
    if has_observed:
        component_rows.append("observed")
    if has_trend:
        component_rows.append("trend")
    if has_seasonal_sum:
        component_rows.append("seasonal")
    # Individual seasonal components (MSTL)
    for s in sorted(seasonal_parts, key=lambda x: int(x.split("_", 1)[1])):
        component_rows.append(s)
    if has_remainder:
        component_rows.append("remainder")

    if not component_rows:
        raise ValueError("No recognizable components found in df (e.g., observed/trend/seasonal/remainder).")

    main_title = title or "Decomposition"

    # ---------- Plotly ----------
    if engine == "plotly":
        import plotly.graph_objects as go
        from plotly.subplots import make_subplots

        # Prepare subplots: one row per component
        fig = make_subplots(
            rows=len(component_rows),
            cols=1,
            shared_xaxes=True,
            vertical_spacing=0.04,
            subplot_titles=component_rows,
        )

        # Helper to add traces for one series across all components
        def _add_series_traces(_uid: str, visible: bool):
            sub = df[df[id_col] == _uid]
            # Add each component row
            for r, comp in enumerate(component_rows, start=1):
                if comp not in sub.columns:
                    continue
                tr = go.Scatter(
                    x=sub[date_col],
                    y=sub[comp],
                    mode="lines",
                    name=f"{_uid}  {comp}",
                    line=dict(width=2),
                    opacity=0.95,
                    visible=visible,
                    connectgaps=False,  # do not connect across NaNs
                )
                # Color code: use palette per series across rows for consistency
                # pick the same color per uid based on its index
                color_idx = ids.index(_uid) % len(PALETTE)
                tr.line.color = PALETTE[color_idx]
                fig.add_trace(tr, row=r, col=1)

        if mode in ("overlay", "facet"):
            # All series visible
            for uid in ids:
                _add_series_traces(uid, visible=True)

            fig.update_layout(
                title=dict(text=main_title, y=0.98),
                legend=dict(yanchor="top", y=0.92),
                height=max(350, 220 * len(component_rows)),
            )
            return _apply_tsforge_style(fig, engine="plotly", context="decomp")

        elif mode == "dropdown":
            # Add all series traces invisible except the first series
            for i, uid in enumerate(ids):
                _add_series_traces(uid, visible=(i == 0))

            # Each series contributes len(component_rows) traces
            traces_per_series = len(component_rows)

            # Build dropdown to toggle which block of traces is visible
            buttons = []
            for i, uid in enumerate(ids):
                vis = [False] * (traces_per_series * n_series)
                start = i * traces_per_series
                for j in range(traces_per_series):
                    vis[start + j] = True
                buttons.append(dict(
                    label=str(uid),
                    method="update",
                    args=[{"visible": vis}]
                ))

            fig.update_layout(
                updatemenus=[{
                    "buttons": buttons,
                    "direction": "down",
                    "x": 1.05, "y": 1.15,
                    "xanchor": "left", "yanchor": "top"
                }],
                title=dict(text=main_title, y=0.98),
                legend=dict(yanchor="top", y=0.92),
                height=max(350, 220 * len(component_rows)),
            )
            return _apply_tsforge_style(fig, engine="plotly", context="decomp")

        else:
            raise ValueError("mode must be 'overlay', 'facet', or 'dropdown'")

    # ---------- Matplotlib ----------
    elif engine == "matplotlib":
        import matplotlib.pyplot as plt

        fig, axes = plt.subplots(len(component_rows), 1, figsize=(10, 2.8 * len(component_rows)), sharex=True)
        axes = np.atleast_1d(axes)

        for ax, comp in zip(axes, component_rows):
            for i, uid in enumerate(ids):
                sub = df[df[id_col] == uid].sort_values(date_col)
                if comp not in sub.columns:
                    continue
                ax.plot(
                    sub[date_col],
                    sub[comp],
                    label=str(uid),
                    color=PALETTE[i % len(PALETTE)],
                    lw=2,
                    alpha=0.95,
                )
            ax.set_title(comp)
            ax.grid(True, linestyle=":", alpha=0.5)

        axes[-1].set_xlabel("Time")
        if len(ids) > 1:
            axes[0].legend(loc="upper left", frameon=False)

        fig.suptitle(main_title, fontsize=14, y=0.98)
        fig.tight_layout()
        return _apply_tsforge_style(fig, engine="matplotlib", context="decomp")

    else:
        raise ValueError("engine must be 'plotly' or 'matplotlib'")
```
